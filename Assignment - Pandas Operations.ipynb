```python
def put_nan_price(column):
  if column == '?' or column == 'n.a':
   return np.nan
  else:
    return float(column)
```

# <b> Problem 10 : Create a new column which stores the number of doors in a car as integers. 

```python
def int_door(x):
  y = 0
  if x == 'one':
    y = 1
  if x == 'two':
    y = 2
  if x == 'three':
    y = 3
  if x == 'four':
    y = 4
  if x == 'five':
    y = 5
  if x == 'six':
    y = 6
  return y
```

```python
# Your code here. Use apply
df['door-as-integer'] = df['num-of-doors'].apply(int_door)

df.tail(20)
```
